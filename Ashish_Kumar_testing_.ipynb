```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

```

```python
dataFrame=pd.read_csv('re_dataset_cancer (2).csv')
dataFrame

data = dataFrame.drop(['Unnamed: 0', 'entry'], axis=1)
data
```

```python
X=data.drop('output',axis=1)
y=data['output']
X
```

```python
X.shape
```

```python
from sklearn.feature_selection import VarianceThreshold
fs=VarianceThreshold(threshold=0.1)
fs.fit_transform(data)
data2=data.loc[:,fs.get_support()]
data2
X2=data2.drop('output',axis=1)
y2=data2['output']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test,y_train,y_test=train_test_split(X2,y2,test_size=0.2,random_state=42,stratify=y2)
```

```python
from lazypredict.Supervised import LazyClassifier
from sklearn.model_selection import train_test_split
X2=data2.drop('output',axis=1)
y2=data2['output']
X_train, X_test,y_train,y_test=train_test_split(X2,y2,test_size=0.2,random_state=42,stratify=y2)
```

```python
from sklearn.metrics import matthews_corrcoef
clf=LazyClassifier(verbose=0,ignore_warnings=True,custom_metric=matthews_corrcoef)
models_train,predictions_train=clf.fit(X_train,X_train,y_train,y_train)
models_test,predictions_test=clf.fit(X_train,X_test,y_train,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=500)
rf.fit(X_train,y_train)
```

```python
y_train_pred=rf.predict(X_train)
y_test_pred=rf.predict(X_test)
```

```python
data2['output']
```

```python
rf.score(X_test,y_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_test_pred)
```

```python
from sklearn.metrics import matthews_corrcoef
matthews_corrcoef(y_test,y_test_pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_test_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_train,y_train_pred))
```

```python
importance=pd.Series(rf.feature_importances_,name='Gini')

feature_names=pd.Series(X2.columns,name='Features')

feature_importance=pd.concat([feature_names,importance],axis=1,names=['Feature','Gini'])

feature_importance
```
